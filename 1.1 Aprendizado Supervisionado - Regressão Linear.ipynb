# 3. Regressão Linear

## 3.1 Divisão dos dados:

```python
import numpy as np
from sklearn.model_selection import train_test_split

features = ['RM'] # average number of rooms per dwelling

X = np.array(housing[features])
Y = np.array(housing['MEDV'])

# divisão dos conjuntos de treino (80%) e teste (20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, random_state = 5487)

pd.core.frame.DataFrame(Y_train)
```

```python
pd.core.frame.DataFrame(Y)
```

## 3.2 Definição da Hipótese:

```python
from sklearn.linear_model import LinearRegression

hipotese = LinearRegression() # y = t0 + t1*x
hipotese.fit(X_train, Y_train) # treinamento do modelo

print(hipotese.intercept_, hipotese.coef_)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (5, 3.5), dpi = 150)

id = 0

plt.ylabel('y')
plt.xlabel('x')

X_plt = X_train.T
Y_plt = Y_train

model_x = np.linspace(3, 8, 100)
model_y = hipotese.intercept_ + hipotese.coef_[0] * model_x

plt.scatter(X_plt, Y_plt)
plt.plot(model_x, model_y, 'r-')
plt.show()
```

## 3.3 Métricas:

https://scikit-learn.org/stable/modules/model_evaluation.html

https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html

```python
Y_pred = hipotese.predict(X_test) # previsão do modelo
Y_pred
```

```python
from sklearn.metrics import r2_score

# avalia o modelo no conjunto de teste
print("r2 =", r2_score(Y_test, Y_pred)) # compara as previsões do modelo com os valores reais
```

# 4. Regressão com várias variáveis

${\displaystyle h(\mathbf {x})=\beta _{0}+\beta _{1}x_{1}+\cdots +\beta _{m}x_{m}}$

${\displaystyle h(\mathbf {x})=\beta_0 +\mathbf {x}^{\mathsf {T}}{\boldsymbol {\beta }}}$


```python
housing.columns
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

features = ['CRIM','ZN','INDUS','CHAS','NO','RM','AGE','DIS','RAD','TAX','PTRATIO', 'B', 'LSTAT']

X = np.array(housing[features])
Y = np.array(housing['MEDV'])

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, random_state = 5487)

pd.core.frame.DataFrame(X).head()
```

```python
from sklearn.linear_model import LinearRegression

hipotese = LinearRegression()
hipotese.fit(X_train, Y_train)

print(hipotese.intercept_, hipotese.coef_)
```

### 💡 Entendendo o Gráfico da Regressão Múltipla

Visualizar uma regressão linear com apenas uma variável (como número de quartos vs. preço) é simples em um gráfico 2D. No entanto, o nosso modelo atual utiliza **13 variáveis simultaneamente**. Como não podemos desenhar um gráfico com 14 dimensões, utilizamos uma projeção matemática para condensar a informação em um plano visualizável.

#### A Matemática da Projeção (`X_train @ hipotese.coef_.T`)
A equação fundamental da regressão múltipla é definida por:
$$h(\mathbf{x}) = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_mx_m$$

Onde:
* $\beta_0$: É o intercepto (`hipotese.intercept_`).
* $\beta_1x_1 + \dots$: É a soma de cada variável (feature) multiplicada pelo seu respectivo peso (coeficiente).

No Python, o operador **`@`** realiza a multiplicação de matrizes (produto escalar). O comando `X_train @ hipotese.coef_.T` multiplica os valores das 13 colunas de cada amostra (casa) pelos seus respectivos pesos aprendidos pelo modelo e soma tudo. O resultado é um **único número** que resume o impacto de todas as características daquela casa (ainda sem o intercepto).

#### Como ler os Eixos e a Linha de Tendência
* **Eixo X (`coef*x`)**: Representa a soma ponderada de todas as variáveis da casa (o "número resumo" calculado com o operador `@`).
* **Eixo Y (`y`)**: Representa o preço real da casa (`Y_train`).
* **A Linha Vermelha (`model_y`)**: É traçada somando o intercepto ao valor do Eixo X (`hipotese.intercept_ + X_plt`). Isso reconstrói exatamente a equação completa da regressão. Ou seja, **a linha vermelha representa a previsão final do modelo**.

**Resumo:** Este gráfico plota o peso acumulado das características (Eixo X) contra os valores reais (Eixo Y). Quanto mais próximos os pontos azuis (valores reais) estiverem grudados na linha vermelha (previsão), menor é o erro e melhor é o ajuste do seu modelo.

```python
(X_train @ hipotese.coef_.T).shape
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (5, 3.5), dpi = 150)

id = 0

plt.ylabel('y')
plt.xlabel('coef*x')

X_plt = X_train @ hipotese.coef_.T
Y_plt = Y_train

model_x = X_plt
model_y = hipotese.intercept_ + X_plt

plt.scatter(X_plt, Y_plt)
plt.plot(model_x, model_y, 'r-')
plt.show()
```

```python
from sklearn.metrics import r2_score

Y_pred = hipotese.predict(X_test)
print("r2 =", r2_score(Y_test, Y_pred))
```

# 5. Melhorias no Modelo

## 4.1 Normalização dos Dados

```python
housing.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Usando as variáveis definidas na seção 3 da regressão múltipla
features_all = ['CRIM','ZN','INDUS','CHAS','NO','RM','AGE','DIS','RAD','TAX','PTRATIO', 'B', 'LSTAT']
```

```python
X_all = np.array(housing[features_all])
Y_all = np.array(housing['MEDV'])

X_train_all, X_test_all, Y_train_all, Y_test_all = train_test_split(X_all, Y_all, train_size = 0.8, random_state = 5487)
```

```python
# Instanciando e aplicando o normalizador
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_all)
X_test_scaled = scaler.transform(X_test_all) # Note que usamos apenas transform() no teste
```

```python
X_train_scaled
```

```python
hipotese_scaled = LinearRegression()
hipotese_scaled.fit(X_train_scaled, Y_train_all)
```

```python
Y_pred_scaled = hipotese_scaled.predict(X_test_scaled)
print("r2 (Normalizado) =", r2_score(Y_test_all, Y_pred_scaled))
```

## 4.2 Tratando Multicolinearidade

```python
# Ao analisar a matriz de correlação, RAD e TAX possuem correlação fortíssima (~0.91)
# Vamos remover 'TAX' para evitar multicolinearidade
features_multi = ['CRIM','ZN','INDUS','CHAS','NO','RM','AGE','DIS','RAD','PTRATIO', 'B', 'LSTAT']
```

```python
X_multi = np.array(housing[features_multi])

# Nova divisão
X_train_multi, X_test_multi, Y_train_multi, Y_test_multi = train_test_split(X_multi, Y_all, train_size = 0.8, random_state = 5487)

# Normalizando os novos dados
X_train_multi_scaled = scaler.fit_transform(X_train_multi)
X_test_multi_scaled = scaler.transform(X_test_multi)
```

```python
hipotese_multi = LinearRegression()
hipotese_multi.fit(X_train_multi_scaled, Y_train_multi)
```

```python
Y_pred_multi = hipotese_multi.predict(X_test_multi_scaled)
print("r2 (Sem TAX - Multicolinearidade) =", r2_score(Y_test_multi, Y_pred_multi))
```

## 4.3 Tratando Outliers

```python
# O dataset de Boston possui um limite artificial (teto) de preços fixado em 50.0 ($50,000)
# Isso cria uma linha horizontal de outliers que prejudica a regressão linear.
# Vamos remover os registros onde MEDV == 50.0

housing_clean = housing[housing['MEDV'] < 50.0]
```

```python
sum(housing['MEDV'] >= 50.0)
```

```python
X_clean = np.array(housing_clean[features_multi]) # Mantendo as melhorias anteriores (sem TAX)
Y_clean = np.array(housing_clean['MEDV'])

# Nova divisão com os dados limpos
X_train_clean, X_test_clean, Y_train_clean, Y_test_clean = train_test_split(X_clean, Y_clean, train_size = 0.8, random_state = 5487)
```

```python
# Normalizando
X_train_clean_scaled = scaler.fit_transform(X_train_clean)
X_test_clean_scaled = scaler.transform(X_test_clean)
```

```python
hipotese_clean = LinearRegression()
hipotese_clean.fit(X_train_clean_scaled, Y_train_clean)
```

```python
Y_pred_clean = hipotese_clean.predict(X_test_clean_scaled)
print("r2 (Sem Outliers de Preço) =", r2_score(Y_test_clean, Y_pred_clean))
```
